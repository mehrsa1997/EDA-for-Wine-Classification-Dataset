The **Wine Dataset** is a widely used dataset for classification and regression problems, particularly in machine learning and statistics. It originates from the UCI Machine Learning Repository and consists of chemical analysis data for wines grown in the same region in Italy but derived from three different cultivars.

### Detailed Characteristics:

1. **Dataset Size**:
   - **178 samples (rows)**: Each represents a wine sample.
   - **13 features (columns)**: Numerical measurements related to the chemical composition of the wine.
   - **1 target variable**: Represents the cultivar (type of wine) to which the sample belongs.

2. **Features**:
   - **alcohol**: Percentage of alcohol content.
   - **malic_acid**: Amount of malic acid in the wine.
   - **ash**: Measurement of ash content in the wine.
   - **alcalinity_of_ash**: Alkalinity of ash (measured in potassium carbonate equivalents).
   - **magnesium**: Magnesium content in parts per million (ppm).
   - **total_phenols**: Total phenolic content.
   - **flavanoids**: Amount of flavonoids, a type of phenolic compound.
   - **nonflavanoid_phenols**: Amount of non-flavonoid phenolic compounds.
   - **proanthocyanins**: A class of flavonoids.
   - **color_intensity**: Measurement of the wine's color intensity.
   - **hue**: The tone or shade of the wine.
   - **od280/od315_of_diluted_wines**: Ratio of absorbance of ultraviolet light at 280 and 315 nm, indicative of phenolic content.
   - **proline**: An amino acid in the wine, typically related to aging.

3. **Target Variable**:
   - **target**: Integer labels (0, 1, 2) representing the type of wine:
     - **0**: Cultivar 1
     - **1**: Cultivar 2
     - **2**: Cultivar 3

4. **Key Properties**:
   - All features are continuous and numerical.
   - No missing values are present in the dataset.
   - The dataset is balanced, with all classes represented reasonably equally.

5. **Use Cases**:
   - **Classification**: Predicting the wine cultivar based on chemical composition.
   - **Feature Analysis**: Identifying which chemical features are most important for distinguishing between cultivars.
   - **Clustering**: Exploring natural groupings in the data.

6. **Statistical Insights**:
   - Features vary widely in their scale. For example:
     - **alcohol**: Around 11.03 to 14.83.
     - **proline**: From 278 to 1680.
     - **hue**: A narrow range of 0.48 to 1.71.
   - This suggests that normalization or scaling is crucial before applying machine learning models.

7. **Limitations**:
   - The dataset is relatively small (178 samples), so care must be taken to avoid overfitting when building models.
   - The dataset focuses on a specific region and set ofy specific machine learning models to this dataset?

## Skewness and Kurtosis Tests ##

**Skewness** and **kurtosis** tests are fundamental for assessing the shape of a distribution and determining its deviation from normality.

#### 1. **Skewness Test**
   - Skewness measures the asymmetry of a distribution:
     - **Symmetric Distribution**: Skewness is close to 0.
     - **Positive Skewness**: Indicates a longer tail on the right (e.g., income distributions where high values stretch the range).
     - **Negative Skewness**: Indicates a longer tail on the left (e.g., age distributions in certain populations).
   - A general rule of thumb:
     - **|Skewness| < 0.5**: Approximately symmetric.
     - **0.5 < |Skewness| < 1**: Moderately skewed.
     - **|Skewness| > 1**: Highly skewed.

#### 2. **Kurtosis Test**
   - Kurtosis measures the \"tailedness\" or peak sharpness of a distribution:
     - **Mesokurtic (Kurtosis = 3)**: Indicates a normal distribution.
     - **Leptokurtic (Kurtosis > 3)**: Distributions with sharper peaks and heavier tails, often indicating outliers.
     - **Platykurtic (Kurtosis < 3)**: Distributions with flatter peaks and lighter tails, indicating fewer extreme values.

#### Why Use These Tests?
- Skewness and kurtosis together provide a comprehensive understanding of distribution shape:
  - **Skewness** highlights symmetry or lack thereof.
  - **Kurtosis** reveals the presence of extreme values or flatness.
- These tests are particularly useful in deciding:
  - Whether to apply transformations (e.g., log, square root) to normalize skewed data.
  - The suitability of statistical methods that assume normality, such as t-tests or ANOVA.

By incorporating these tests, data analysts can address deviations from normality effectively, improving the reliability of downstream analyses and modeling.

## Outliers Detection Using the IQR Method ##

The Interquartile Range (IQR) method is a robust statistical technique to identify outliers in a dataset. It uses the spread of the middle 50% of the data to set boundaries, outside which values are considered outliers.

### Enhancing Outlier Detection with Data Transformation

Outlier detection methods like the Interquartile Range (IQR) technique assume a relatively symmetric distribution to accurately identify extreme values. For skewed data, this assumption can lead to misclassification of legitimate data points as outliers or failure to detect true anomalies. To address this, data transformation is a crucial preprocessing step to improve the effectiveness of the IQR method.


### **Why Transform the Data Before IQR?**
1. **Handle Skewness**:
   - Skewed distributions distort the calculation of quartiles, leading to biased IQR ranges.
   - Transformations reduce skewness, making the data more symmetric and the IQR boundaries more reliable.

2. **Preserve Data Integrity**:
   - Without transformation, legitimate high or low values in skewed data may be incorrectly flagged as outliers.
   - Proper transformations align the data closer to its central tendency, ensuring valid points are not treated as anomalies.

3. **Enable Better Interpretation**:
   - Post-transformation, data points are often easier to interpret and compare within the context of their adjusted range.

### **Benefits of Preprocessing with Transformations**
1. **Enhanced Accuracy**: Transformations align the data with the assumptions of IQR, reducing false positives or negatives in outlier detection.
2. **Increased Reliability**: Properly adjusted data ensures that the identified outliers genuinely deviate from the central trend.
3. **Flexibility**: Different transformation methods cater to varying degrees and types of skewness, making the process adaptable to diverse datasets.


### **Conclusion**
Transforming data before applying the IQR method is essential for skewed distributions. It enhances the robustness of outlier detection by addressing asymmetry and ensuring that the quartile boundaries accurately represent the dataset. Among the transformation methods, log, square root, and Box-Cox transformations provide flexible solutions for a wide range of skewed data scenarios. Incorporating this preprocessing step significantly improves the quality of outlier analysis and downstream statistical or machine learning processes.

```python
data_before_cleaning=wine_data.copy()
```

```python
wine_data = np.log(wine_data + 1)  # Avoid log(0)
```

```python
# Identify outliers using IQR method and return their indices
outlier_indices = []

for feature in wine_data.columns:
    # Calculate Q1, Q3, and IQR for each feature
    Q1 = wine_data[feature].quantile(0.25)
    Q3 = wine_data[feature].quantile(0.75)
    IQR = Q3 - Q1
    
    # Define lower and upper bounds
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Identify indices of outliers for the current feature
    feature_outliers = wine_data[(wine_data[feature] < lower_bound) | (wine_data[feature] > upper_bound)].index
    outlier_indices.extend(feature_outliers)

# Get unique indices of outliers across all features
unique_outlier_indices = list(set(outlier_indices))
unique_outlier_indices
```

## build machine learning model ##


Below is the complete and ordered code. I implemented the use of a Pipeline to reduce rows and speed up calculations and processes. (Previously, to explain every choice and every step, it was not possible for me to use it)

```python
import numpy as np
import pandas as pd
from sklearn import datasets
from sklearn.model_selection import train_test_split as split
from sklearn.svm import SVC
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score as accuracy

#loading the dataset
wine = datasets.load_wine()
X, y = wine.data, wine.target

#separating training and test data
X_train, X_test, y_train, y_test = split(X, y, test_size=0.3, shuffle=True, random_state=0, stratify=y)


#creating the pipeline containing the scaler, pca and the model
pipe = Pipeline([("scaler", MinMaxScaler()),
                 ("svc", SVC())])

#preparing the values of hyperparameters to be validated
parameters = [{"svc__kernel": ["linear"], "svc__C": [0.01, 0.1, 1, 10, 100]},
              {"svc__kernel": ["rbf"], "svc__C": [0.01, 0.1, 1, 10, 100], "svc__gamma": [0.01, 0.1, 1, 10, 100]},
              {"svc__kernel": ["poly"], "svc__C": [0.01, 0.1, 1, 10, 100], "svc__degree": np.arange(1,5,1)}]

#set the number of subset to be created for validation
crossval= StratifiedKFold(n_splits=5, shuffle=True, random_state=0)

#finding the best values to assign to hyperparameters
clf = GridSearchCV(pipe, param_grid=parameters, cv=crossval, n_jobs=-1)

#training the model with the best hyperparameters found (GridSearchCV applies them automatically)
#and I evaluate the accuracy on training and test sets
clf.fit(X_train, y_train)

pred_train = clf.best_estimator_.predict(X_train)
pred_test = clf.best_estimator_.predict(X_test)
print(f"Best parameters are: {clf.best_params_}, with a score of {round(clf.best_score_,3)}")
print(f"Accuracy on training set is: {round(accuracy(y_train, pred_train), 3)}")
print(f"Accuracy on test set is : {round(accuracy(y_test, pred_test), 3)}")
```

The accuracy obtained is very good but I still print a classification report and a confusion matrix to get a clearer view of the errors

A Classification report is used to measure the quality of predictions from a classification algorithm. How many predictions are True and how many are False. More specifically, True Positives, False Positives, True negatives and False Negatives are used to predict the metrics of a classification report. Precision is the ability of a classifier not to label an instance positive that is actually negative. Recall is the ability of a classifier to find all positive instances The F1 score is a weighted harmonic mean of precision and recall

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, pred_test, labels=[0,1,2]))
```

In this model recall is equal to one for all the classes( except class 0 but with a great result too) and precision is equal to one for all the classes( except class 1 but with another great result). Finally take a look at the confusion matrix

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# Compute confusion matrix
cm = confusion_matrix(y_test, pred_test)

# Display confusion matrix
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=[0, 1, 2])
disp.plot(cmap=plt.cm.Blues)
plt.show()
```

We see that only one prediction was wrong (a sample from class 0 was assigned to class 1). I can therefore be satisfied with the result achieved.
